# Proyecto de ACUS 220: Desarrollo de bot de discord
### Segundo Hito de Entrega
**Autor:** Rodrigo González

## Recordando el objetivo general

- Desarrollo de un bot usando python, el cual podrá manipular audios de diferentes formas, tales como:

    - Cambio de volumen, saturación, velocidad, etc.
    - Creación de audios mediante TTS
    - Efectos varios de sonido

Busco proporcionar una herramienta a la mano para todos los usuarios de la plataforma, los cuales podrán evitar sitios web de dudosa calidad y confianza para poder editar sus audios.

No se requiere un conocimiento de edición de audios para usar, pues busco crear comando simples y autoexplicativos. De todos modos, el bot tendrá un comando específico donde se indiquen las instrucciones.

## Avance del hito 2

- Desde el hito 2, he añadido diferentes funcionalidades, las cuales son:

    - Cambio de volumen, velocidad , modo reversa, efecto fade y recortes de audios
    - Generación de audios usando TTS (text to speech) con el comando **!generartts**
    - Reproducción de texto a audio en tiempo real durante llamada de Discord con **!h** y **!tts**

## Propuestas para implementaciones para el hito 3

- Para el tercer hito, me gustaría implementar otras funcionalidades y objetivos:

    - Hacer deploy del bot, para que pueda usarse desde un servidor externo
    - Crear una página de documentación en línea con los comandos
    - Implementar un sistema de recuerdo: Si el usuario desea que todos sus mensajes sean TTS, añadirlo a una lista para que el bot siga este comando
    - [si es que consigo una API que no cueste dinero] Implementación de voces IA para el TTS en tiempo real
